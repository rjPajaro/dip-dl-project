# Image Data Cleaning

```python
import pandas as pd
```

### RFMiD Data Cleaning

```python
rfmid = pd.read_csv('../data/metadata/RFMiD/RFMiD_Training_Labels.csv')

# rfmid w/o other diseases apart from amd
rfmid_pure_amd = pd.read_csv('../data/metadata/RFMiD/RFMiD_AMD-wo-other-diseases.csv')
rfmid.head()
```

```python
rfmid_pure_amd.head()
```

```python
rfmid_pure_amd_IDs = rfmid_pure_amd['ID'].to_list()
print('rfmid_pure_amd_IDs =', len(rfmid_pure_amd_IDs))
print('rfmid_pure_amd =', len(rfmid_pure_amd))
```

```python
# rfmid w/ other diseases apart from amd
rfmid_amd_others = rfmid

for i in range(len(rfmid_amd_others)):
    for j in range(len(rfmid_pure_amd_IDs)):
        if rfmid_amd_others['ID'][i] == rfmid_pure_amd_IDs[j]:
            rfmid_amd_others = rfmid_amd_others.drop(i)
            break

rfmid_amd_others = rfmid_amd_others.loc[rfmid_amd_others['ARMD'] == 1] .reset_index(drop=True)
#rfmid_amd_others.to_csv('../data/RFMiD/Training_Set/Training_Set/RFMiD_AMD-w-other-diseases.csv')
rfmid_amd_others.head()
```

```python
rfmid_normals = rfmid.loc[rfmid['Disease_Risk'] == 0]
#rfmid_normals.to_csv('../data/RFMiD/Training_Set/Training_Set/RFMiD_normals.csv')
rfmid_normals.head()
```

```python
print('Total of rfmid_amd_others =', len(rfmid_amd_others))
print('Total of rfmid_amd_w/o_others =', len(rfmid_pure_amd))
print('Total of rfmid normals = ', len(rfmid_normals))
```
