### RFMiD Data Cleaning

```python
rfmid = pd.read_csv('../data/raw-data/RFMiD/Evaluation_Set/Evaluation_Set/RFMiD_Validation_Labels.csv')

# rfmid w/o other diseases apart from amd
rfmid_pure_amd = pd.read_excel('../data/raw-data/RFMiD/Evaluation_Set/Evaluation_Set/RFMiD_PureAMD_Validation_Labels.xlsx')
rfmid.head()
```

```python
rfmid_pure_amd.head()
```

```python
rfmid_pure_amd_IDs = rfmid_pure_amd['ID'].to_list()
print('rfmid_pure_amd_IDs =', len(rfmid_pure_amd_IDs))
print('rfmid_pure_amd =', len(rfmid_pure_amd))
```

```python
# rfmid w/ other diseases apart from amd
rfmid_amd_others = rfmid

for i in range(len(rfmid_amd_others)):
    for j in range(len(rfmid_pure_amd_IDs)):
        if rfmid_amd_others['ID'][i] == rfmid_pure_amd_IDs[j]:
            rfmid_amd_others = rfmid_amd_others.drop(i)
            break

rfmid_amd_others = rfmid_amd_others.loc[rfmid_amd_others['ARMD'] == 1] .reset_index(drop = True)
#rfmid_amd_others.to_csv('../data/metadata/RFMiD/Training_Set/Training_Set/RFMiD_AMD-w-other-diseases.csv')
rfmid_amd_others.head()
```

```python
rfmid_normals = rfmid.loc[rfmid['Disease_Risk'] == 0]
#rfmid_normals.to_csv('../data/RFMiD/Training_Set/Training_Set/RFMiD_normals.csv')
rfmid_normals.head()
```

```python
print('Total of rfmid_amd_others =', len(rfmid_amd_others))
print('Total of rfmid_amd_w/o_others =', len(rfmid_pure_amd))
print('Total of rfmid normals = ', len(rfmid_normals))
```

### ODIR 

```python
L_odir_pure_amd = pd.read_csv('../data/metadata/ODIR/L-wo-others.csv')
R_odir_pure_amd = pd.read_csv('../data/metadata/ODIR/R-wo-others.csv')
```

```python
odir_pure_amd = L_odir_pure_amd.merge(R_odir_pure_amd)
odir_pure_amd
```

```python
L_odir_amd_others = pd.read_csv('../data/metadata/ODIR/L-w-others.csv')
R_odir_amd_others = pd.read_csv('../data/metadata/ODIR/R-w-others.csv')
```

```python
odir_amd_others = L_odir_amd_others.merge(R_odir_amd_others, how = 'outer')
odir_amd_others.head()
```

```python
odir_normals = pd.read_csv('../data/metadata/ODIR/ODIR-normal.csv')
odir_normals
```

# Data Filtering

```python
#ODIR Normal
data_folder = '../data/ODIR/ODIR-5K/ODIR-5K/Training Images'
select_folder = '../data/filtered-data/ODIR/Normal'

right_normal_img = odir_normals['Right-Fundus']
left_normal_img = odir_normals['Left-Fundus']

sel_img = right_normal_img.append(left_normal_img, ignore_index = True)
sel_img[1]

def select_img():
        for file in sel_img:
                shutil.copy(f'{data_folder}/{file}', f'{select_folder}/{file}')

select_img()
```

```python
#ODIR Pure AMD
#data_folder = '../data/ODIR/ODIR-5K/Training Images'
select_folder = '../data/filtered-data/ODIR/Pure AMD'

right_normal_img = odir_pure_amd['Right-Fundus'].dropna()
left_normal_img = odir_pure_amd['Left-Fundus'].dropna()

sel_img = right_normal_img.append(left_normal_img, ignore_index = True)
sel_img[1]

select_img()
```

```python
#ODIR AMD w others
data_folder = '../data/ODIR/ODIR-5K/ODIR-5K/Training Images'
select_folder = '../data/filtered-data/ODIR/AMD w others'

right_normal_img = R_odir_amd_others['Right-Fundus'].dropna()
left_normal_img = L_odir_amd_others['Left-Fundus'].dropna()

sel_img = right_normal_img.append(left_normal_img, ignore_index = True)
sel_img[1]

select_img()
```

```python
#RFMiD Normal
data_folder = '../data/raw-data/RFMiD/Evaluation_Set/Evaluation_Set/Validation'
select_folder = '../data/preprocessing/filtered-data/RFMiD/Validation/Normal'

sel_img = rfmid_normals['ID']

def select_img_RFMiD():
        for file in sel_img:
                shutil.copy(f'{data_folder}/{file}' + '.png' , f'{select_folder}/{file}' + '.png')

select_img_RFMiD()
```

```python
#RFMiD Pure AMD
#data_folder = '../data/RFMiD/Training_Set/Training'
select_folder = '../data/preprocessing/filtered-data/RFMiD/Validation/Pure AMD'

sel_img = rfmid_pure_amd['ID']

select_img_RFMiD()
```

```python
#RFMiD AMD w others
#data_folder = '../data/RFMiD/Training_Set/Training'
select_folder = '../data/preprocessing/filtered-data/RFMiD/Validation/AMD w others'

sel_img = rfmid_amd_others['ID']

select_img_RFMiD()
```
